```python
def priceTracer():
    import requests
    from bs4 import BeautifulSoup
    url= 'https://finance.yahoo.com/quote/BTC-USD?p=BTC-USD&.tsrc=fin-srch'
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'lxml')
    price = soup.find_all('div',{'class':'D(ib) Mend(20px)'})[0].find('span').text
    return price
```

```python
while True:
    print('The current price of Bitcoin is ' +priceTracer())
```
